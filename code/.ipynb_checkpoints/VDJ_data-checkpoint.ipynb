```python
import scanpy as sc
import scirpy as ir
import muon as mu
import anndata as ad
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm as mpl_cm
from cycler import cycler
```

```python
rawDataPath = "/mnt/iacchus/joe/raw_data/"
processedDataPath = "/mnt/iacchus/joe/processed_data/"
```

```python
adata_tcr_WT = ir.io.read_10x_vdj(
    f"{rawDataPath}Thymus/Xian/cellranger_output/CCR6_WT_B_cells_5v2BCR_cellranger/filtered_contig_annotations.csv"
)
adata_tcr_KO = ir.io.read_10x_vdj(
    f"{rawDataPath}Thymus/Xian/cellranger_output/CCR6_KO_B_cells_5v2BCR_cellranger/filtered_contig_annotations.csv"
)
sc.pp.subsample(adata_tcr_KO, n_obs=adata_tcr_WT.shape[0])
adata_tcr = ad.concat(
    [adata_tcr_WT, adata_tcr_KO],
    label="batch",
    keys=["WT", "CCR6_KO"]
)
adata_tcr.obs_names_make_unique()
```

```python
adata = sc.read(f"{processedDataPath}/Thymic_B_cells/h5ad_files/CCR6_WT_KO_B_cells.h5ad")
adata = adata[adata.obs_names.isin(adata_tcr.obs_names)]
adata_tcr = adata_tcr[adata_tcr.obs_names.isin(adata.obs_names)]
```

```python
mudata = mu.MuData({"gex": adata, "airr": adata_tcr})
```

```python
ir.pp.index_chains(mudata)
ir.tl.chain_qc(mudata)
```

```python
_ = ir.pl.group_abundance(
    mudata, groupby="airr:receptor_subtype", target_col="gex:leiden_scVI"
)
```

```python
_ = ir.pl.group_abundance(mudata, groupby="airr:chain_pairing", target_col="gex:leiden_scVI")
```

```python
print(
    "Fraction of cells with more than one pair of BCRs: {:.2f}".format(
        np.sum(
            mudata.obs["airr:chain_pairing"].isin(
                ["extra VJ", "extra VDJ", "two full chains", "multichain"]
            )
        )
        / mudata["airr"].n_obs
    )
)
```
